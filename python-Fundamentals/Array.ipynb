# Second Largest

```python
def sec_largest(arr):
    large=arr[0]
    seclarge = float('-inf')
    
    for i in range(len(arr)):
        if arr[i]>large :
            seclarge=large
            large=arr[i]
        elif arr[i]>seclarge and arr[i]!=large:
            seclarge=arr[i]
        
    print(seclarge)

sec_largest([2,6,1,5])
sec_largest([5, 5, 3])
sec_largest([1, 2])
```

# Find even no


```python
def count_evens(arr):
    # count how many even numbers in array
    n = 0
    for i in range (len(arr)) :
        if arr[i] % 2 == 0 :
            n=n+1
    return n

print(count_evens([1,2,3,4,5,6]))  # Expected: 3
print(count_evens([1,3,5]))         # Expected: 0
```

## Sum of array

```python
def array_sum(arr):
    # return sum of all elements
    # don't use sum() builtin
    sum =0
    for i in range (len(arr)) :
        sum=sum+ arr[i]
    return sum

print(array_sum([1,2,3,4,5]))  # Expected: 15
print(array_sum([10,20]))       # Expected: 30
```

# Find largest number in an array

```python
def find_largest(arr):
    # return the largest element
    # don't use max() builtin
    largest=arr[0]
    for i in range (len(arr)) :
        if arr[i]>largest:
            largest=arr[i]
    return largest

print(find_largest([3,7,1,9,4]))  # Expected: 9
print(find_largest([5,5,5]))       # Expected: 5
```

# min ele

```python
def find_min(arr):
    minimum=arr[0]
    for i in range (len(arr)) :
        if arr[i]<min:
            minimum= arr[i]
    return minimum 

print(find_min([3,7,1,9,4]))  # Expected: 1
print(find_min([5,5,5]))       # Expected: 5
```

# Count element greater than x

```python
def count_greater(arr, x):
    count=0
    for i in range (len(arr)) :
        if arr[i] > x :
            count = count +1
    return count

print(count_greater([1,5,3,8,2], 4))  # Expected: 2 (5 and 8)
print(count_greater([1,2,3], 5))       # Expected: 0
```

# Replace negatives with zero

```python
def replace_negatives(arr):
    for i in range (len (arr)) :
        if arr[i] < 0:
            arr[i]=0
    return arr

print(replace_negatives([1,-2,3,-4,5]))  # Expected: [1,0,3,0,5]
print(replace_negatives([-1,-2,-3]))      # Expected: [0,0,0]
```

# Replace negatives with zero without changing the given array

```python
def replace_negatives(arr):
    result = []
    for i in range(len(arr)):
        if arr[i] < 0:
            result.append(0)
        else:
            result.append(arr[i])
    return result
```

# Count occurrences of a target

```python
def count_occurrences(arr, target):
    #pass
    count =0
    for i in range (len(arr)) :
        if arr[i] == target :
            count = count+1
    return count

print(count_occurrences([1,2,2,3,2,4], 2))  # Expected: 3
print(count_occurrences([1,2,3], 5))         # Expected: 0
```

# Find sum of even numbers only

```python
def sum_evens(arr):
    #pass
    total = 0
    for i in range (len(arr)) :
        if arr[i] % 2 == 0 :
            total= total + arr[i]
            
    return total

print(sum_evens([1,2,3,4,5,6]))  # Expected: 12 (2+4+6)
print(sum_evens([1,3,5]))          # Expected: 0
```

# Reverse array in place (no new array, no slicing)

```python
def reverse_in_place(arr):
    #pass
    start = 0
    end= len(arr) -1
    
    while (start< end):
        arr[start], arr[end] = arr[end] , arr[start]
        start +=1
        end -=1
    return arr       

arr = [1,2,3,4,5]
reverse_in_place(arr)
print(arr)  # Expected: [5,4,3,2,1]
```

# Find missing number

```python
def find_missing(arr):
    # arr contains n-1 numbers from 1 to n
    # one number is missing, find it
    n=len(arr)+1
    total=n*(n+1)//2
    #print(total)
    arrtotal=0
    for i in range(len(arr)):
        arrtotal=arrtotal+arr[i]
    #print (arrtotal)
    return total-arrtotal
    

print(find_missing([1,2,4,5,6]))   # Expected: 3
print(find_missing([1,2,3,4,6]))   # Expected: 5
```

# Remove Dublicates

```python
def remove_duplicates(arr):
    # return array with duplicates removed
    # maintain original order
    #pass
    result=[]
    for i in range (len(arr)):
        if arr[i] not in result:
            result.append(arr[i])
    return result

print(remove_duplicates([1,2,2,3,3,4]))  # Expected: [1,2,3,4]
print(remove_duplicates([5,5,5,5]))       # Expected: [5]
```

# Two sum (no hashmap)

```python
def two_sum(arr, target):
    # find two numbers that add up to target
    # return their indexes
    #pass
    for i in range (len(arr)):
        for j in range (i+1, len(arr)):
            if arr[i] + arr[j] == target:
                return i,j

print(two_sum([2,7,11,15], 9))   # Expected: [0,1] (2+7=9)
print(two_sum([1,3,5,6], 8))     # Expected: [1,2] (3+5=8)
```
