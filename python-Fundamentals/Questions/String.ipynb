# Reverse A String

```python
def reverse_string(s):
    # don't use s[::-1]
    #pass
    p=[]
    for i in range(len(s)-1-1,-1):
        p.append(s[i])
        return "".join(p)   # join with nothing between each character.


print(reverse_string("hello"))   # Expected: "olleh"
print(reverse_string("neha"))    # Expected: "ahan"
```

# Check if string is palindrome

```python
def is_palindrome_string(s):
    start = 0
    end = len(s) - 1
    
    while start < end:
        if s[start] != s[end]:
            return False
        start += 1
        end -= 1
    
    return True

print(is_palindrome_string("racecar"))  # Expected: True
print(is_palindrome_string("hello"))    # Expected: False
```

# Count vowels in a string
