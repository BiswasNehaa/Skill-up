# Reverse A String

```python
def reverse_string(s):
    # don't use s[::-1]
    #pass
    p=[]
    for i in range(len(s)-1-1,-1):
        p.append(s[i])
        return "".join(p)   # join with nothing between each character.


print(reverse_string("hello"))   # Expected: "olleh"
print(reverse_string("neha"))    # Expected: "ahan"
```

# Check if string is palindrome

```python
def is_palindrome_string(s):
    start = 0
    end = len(s) - 1
    
    while start < end:
        if s[start] != s[end]:
            return False
        start += 1
        end -= 1
    
    return True

print(is_palindrome_string("racecar"))  # Expected: True
print(is_palindrome_string("hello"))    # Expected: False
```

# Count vowels in a string

```python
def count_vowels(s):
    #pass
    count=0
    for i in range(len(s)):
        #if(s[i]=='a' or s[i]=='e' or s[i]=='i' or s[i]=='o' or s[i]=='u' or s[i]=='A' or s[i]=='E' or s[i]=='I' or s[i]=='O' or s[i]=='U' ):
        if s[i].lower() in "aeiou":   
            count = count +1
    return count

print(count_vowels("hello"))        # Expected: 2
print(count_vowels("aeiou"))        # Expected: 5
print(count_vowels("rhythm"))       # Expected: 0
```

# Count Characters

```python
def character_count(s):
    # Return a dictionary containing the frequency of each character.
    # pass
    count={}
    
    for ch in s:
        if ch in count:
            count[ch]+=1
        else:
            count[ch]=1
    
    return count


print(character_count("banana"))   # Expected: {'b': 1, 'a': 3, 'n': 2}
print(character_count("apple"))    # Expected: {'a': 1, 'p': 2, 'l': 1, 'e': 1}
```

# Remove Space

```python
def remove_spaces(s):
    # Remove all spaces from the string.
    # Don't use replace().
    # pass
    ans=[]
    for i in s:
        if i==" ":
            continue
        else:
            ans.append(i)

    return "".join(ans)

print(remove_spaces("I love Python"))   # Expected: "IlovePython"
print(remove_spaces("hello world"))     # Expected: "helloworld"
```

# First Non-Repeating Character

```python
def first_unique(s):
    # Return the first non-repeating character.
    # If none exists, return -1.
    # pass
    for ch in s:
        if s.count(ch)==1:
            return ch
    return -1


print(first_unique("aabbcdde"))   # Expected: "c"
print(first_unique("aabbcc"))     # Expected: -1
```
