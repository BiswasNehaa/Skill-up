# Default arguments: 
=> A default argument is a parameter that assumes a predefined value if no value is provided in the function cal
=> Order of the arguments matters

```python
def add(a=6,b=10):
    c=a+b
    print(c)

add()  #default arguments are used
add(1)  # we are passing a value and b is default
add(a=2) # we are passing a value and b is default
add(b=4)  #we are passing b value and a is default
```

# Keyword Arguments:
=> Allow you to pass values to a function by explicitly stating the parameter name followed by an equals sign (name=value). 
=> This makes the order of the arguments completely irrelevant during the function call.

```python
def sub(a,b):
    c=a-b
    print(c)

sub(4,3)
sub(b=4,a=3)
```

# Required Arguments:
=> parameters defined in a function signature that do not have default values assigned to them. 
=> When you call that function, you are mandatory to pass values for these arguments. 
=> Leaving them out will immediately cause Python to raise a TypeError

```python
def add(a,b=1):
    c=a+b
    print(c)

add(a=3,b=2)
add(5)
#add() #error
```

# Variable-Length Arguments:
=> allow a function to accept an arbitrary number of inputs.

# *args:
=> for non keyword arguments
=> tuples

# **args:
=> Keyword arguments
=> dictionary

```python
# *args : Non-keyword arguments
def avg (*num):   # *args
    sum=0
    for i in num:
        sum=sum+1
    print(sum)
    print (type(num))  #num is of type tuple

avg(1,2,3,4,5,6)



# **args : Keyword arguments

def about(**Desc):
    for key,value in Desc.items():
        print(f"{key}:{value}")
    print(type(Desc))

about(name="Alice",age=11,lname='abc')
```

## Oneshot


A function is a reusable block of code. You define it once, call it many times.

# Basic function
def greet(name: str) -> str:
    return f"Hello {name}"

# Default argument
def greet(name: str = "Neha") -> str:
    return f"Hello {name}"

# *args = multiple positional args (tuple inside)
def add(*args):
    return sum(args)
add(1, 2, 3)  # → 6

# **kwargs = multiple keyword args (dict inside)
def show(**kwargs):
    for k, v in kwargs.items():
        print(k, v)
show(name="Neha", city="Bangalore")



Remember
→ *args  gives you a tuple of extra positional values
→**kwargs gives you a dict of extra keyword values
