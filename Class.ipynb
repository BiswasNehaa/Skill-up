# Constructor:
=> Every constructor in Python is defined using the def __init__(self) syntax. 
=> The first argument, self, is a mandatory reference to the current instance of the object being created.

```python
class person:
    def __init__(self, name, occ):   #parameterized constructor
        self.name=name
        self.occ=occ
    
    def info(self):   # Method inside a class always have self parameter
        print(f"Hey {self.name} is a  {self.occ}")

a=person("Neha","student")
#a.info()
print(a.info())  #will return none as no return statement .. default none
```

# Class Variable VS Instance Variable
Class Variable: A variable declared inside the class but outside methods and shared by all objects.
Instance Variable: A variable declared using self inside methods (usually __init__) and unique to each object.

```python
class Employee:
    
    #class var , shared my all
    company_name = "abc"
    no_of_emp=0
    
    def __init__(self,name,role):
        
        #instannce var.. unique for all
        self.name=name
        self.role=role
        
        self.no_of_emp +=1
        
emp1=Employee("neha","stud")
emp2=Employee("adf","sde")

# access instance Var
print(emp1.name)

#access Class var
print(emp1.company_name)
print(Employee.company_name)
```

# Instance Method
=> An instance method is the default type of method you write inside a Python class. It passes the individual object implicitly as self


# Class Method
A class method belongs to the class itself rather than any individual object. It uses the @classmethod decorator and accepts cls as its first argument.

```python
class Student:
    
    def __init__(self,name,gpa):
        self.name=name  #Instance Var
        self.gpa=gpa    #Instance Var
        
    def info(self):    #instance method
        print(f"Hiii I am {self.name} having {self.gpa}")
        
    @classmethod
    def change_school(cls,name):  #class Method
        cls.name=name
        print (f"new school name {cls.name}")
                
stud1=Student("ABC",7.8)
stud1.info()  # To call instance method use the obj
Student.change_school("ABC")    # To call class method use class name
```

# Inheritance :
 is a core Object-Oriented Programming (OOP) mechanism that allows a child class (subclass) to acquire all the attributes and methods of a parent class (superclass). 


# Override :
occurs when a child class defines a method with the exact same name and signature as a method inherited from its parent class

# Super() :
function is a built-in function used to access and call methods from a parent or sibling class.


```python
class Human:
    
    def eat(self):
        print("I can eat")
    
    def work(self):
        print("I can work")

class Male (Human):  #child class
    def run(self):
        print("I am a male and can run")
        
    def work(self):     #Method Overriding
        print("I code")
        
    def eat(self):   
        super().eat()     # accessing the parent method
        print("And i love Biryani")
        

male1=Male()     # Accessing the methods of parent class
male1.run()
male1.work()
male1.eat()

    
```

```python
class Human:
    
    def __init__(self,no_nose):
        self.no_nose=no_nose
        
    def eat(self):
        print("I can eat")
    
    def work(self):
        print("I can work")

class Male (Human):  #child class
    
    def __init__(self,no_nose, name):
        super().__init__(no_nose)
        self.name=name
        
    def run(self):
        print("I am a male and can run")
        
    def work(self):     #Method Overriding
        print("I code")
        
    def eat(self):   
        super().eat()     # accessing the parent method
        print("And i love Biryani")
        
    def info(self,name,no_nose):
        print(f"Hiiii I am {self.name} and have {self.no_nose} nose")

male1=Male(1,"abc")     # Accessing the methods of parent class
male1.run()
male1.work()
male1.eat()
male1.no_nose   # Aceesing parent instance
male1.info("abcd",1)


    
```

# Define a class
class Document:
    def __init__(self, title: str, content: str):
        self.title = title        # instance variable
        self.content = content

    def summary(self) -> str:
        return f"{self.title}: {self.content[:50]}"

# Create an object (instance)
doc = Document("AI Notes", "RAG is retrieval augmented generation")
print(doc.summary())

# Inheritance — child gets parent's methods
class PDFDocument(Document):
    def __init__(self, title: str, content: str, pages: int):
        super().__init__(title, content)  # call parent __init__
        self.pages = pages
