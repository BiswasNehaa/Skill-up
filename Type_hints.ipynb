# Optional :
An optional type hint in Python indicates that a variable, argument, or return value can either be a specific data type or None

# Any :
In Python, the Any type hint indicates that a variable, parameter, or return value can hold any datatype

# Sequence :
Using a generic Sequence instead of a specific type like list allows your functions to accept any ordered collection that supports length checking and indexing, such as a list, tuple, or str.

```python
from typing import Optional , Any , Sequence

def Hi (a : Optional[bool]=False) :
    pass

Hi()


def foo ( a: Any) -> Any :
    pass

foo()


def foo2(a : Sequence) :
    pass

foo2 ([1 ,2 ,3])
foo2 ((1, 2))
foo2 ("Hello")
```

# Callable :
To type hint a function, lambda, method, or class in Python, use

syntax :
Callable[[ParamType1, ParamType2], ReturnType]


```python
from typing import Callable

def add(x : int , y: int) -> float :
    return x + y
 
def foo (func : Callable[[int, int], float]) -> None :      # [int, int] : Add func param type and float is add func return type
    add(2,3)
    
foo(add)
print(foo(add))
```

## Note:

Type hints tell you what type a variable should be. Python doesn't enforce them at runtime — but Pydantic does. This is exactly why your resume has Pydantic on it.

from typing import List, Dict, Optional, Union

# Basic hints
name: str = "Neha"
age: int = 21
score: float = 8.6
is_active: bool = True

# Collections
topics: List[str] = ["RAG", "FAISS", "FastAPI"]
scores: Dict[str, int] = {"math": 90, "cn": 85}

# Optional = value OR None
def get_user(id: int) -> Optional[str]:
    return "Neha" if id == 1 else None

# Union = one of multiple types
def process(value: Union[str, int]) -> str:
    return str(value)

# Full example with Pydantic-style class
from pydantic import BaseModel

class UserQuery(BaseModel):
    question: str
    max_results: int = 5
    filter_tag: Optional[str] = None

Remember
→ Optional[str] means the value can be a string OR None
→ List[str] means a list where every item is a string
