Read your doc and convert it to raw text . Could be a PDF, CSV or text file.

```python
import langchain
print(langchain.__version__)
```

```python
from langchain_community.document_loaders import PyPDFLoader

# Load a PDF
loader= PyPDFLoader("10 RAG Architectures.pdf")
documents= loader.load()

# Each doc has:
print(documents[0].page_content)
print(documents[0].metadata)
```


# What is a Document object?
→LangChain's Document has two things: page_content (the text) and metadata (source file, page number etc)
→metadata is useful for citing sources — "this answer came from page 3"


Q: What is a LangChain Document object?

A: It has two fields — page_content which holds the actual text, and metadata which holds information like source file and page number.
