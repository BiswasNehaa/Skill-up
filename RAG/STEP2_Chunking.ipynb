# Why chunk? 
A 100-page PDF is too large to embed as one piece. We split it into smaller overlapping chunks so each chunk fits in context and is specific enough to be useful.

```python
# Load Doc

from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("10 RAG Architectures.pdf")
documents=loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter= RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks=splitter.split_documents(documents)
print(f"total chunks: {len(chunks)}")

```


# Key concepts
→chunk_size:
how large each chunk is. Too large = too much irrelevant info. Too small = loses context.

→chunk_overlap:
— how many characters are shared between consecutive chunks. Prevents important info from being cut at chunk boundaries.

→RecursiveCharacterTextSplitter
— tries to split on paragraphs first, then sentences, then words. Keeps text natural.



Q: Why do we chunk documents and what is chunk overlap?
A: We chunk because large documents can't be embedded as one piece and would exceed context window. Chunk overlap ensures important information at chunk boundaries isn't lost — consecutive chunks share some text.
